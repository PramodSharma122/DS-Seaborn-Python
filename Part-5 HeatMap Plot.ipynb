# HeatMap Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var=np.linspace(1,10,20).reshape(4,5)
var
```

```python
sns.heatmap(var)
```

```python
sns.heatmap(var,vmin=0,vmax=12)
```

```python
sns.heatmap(var,vmin=0,vmax=12,cmap="gist_heat")
```

```python
sns.heatmap(var,vmin=0,vmax=12,annot=True)
```

```python
var1=np.linspace(1,10,10).reshape(2,5)
var1
```

```python
sns.heatmap(var1,vmin=0,vmax=10,cmap="PuOr",annot=True)
```

```python
ar=np.array([["a0","a1","a2","a3","a4"],["b0","b1","b2","b3","b4"]])
sns.heatmap(var1,vmin=0,vmax=10,cmap="PuOr",annot=ar,fmt="s")
```

```python
y={"fontsize":20,"color":"r"}
sns.heatmap(var,vmin=0,vmax=10,annot=True,annot_kws=y)
```

```python
y={"fontsize":20,"color":"r"}
sns.heatmap(var,vmin=0,vmax=10,annot=True,annot_kws=y,linewidth=10)
```

```python
y={"fontsize":20,"color":"r"}
sns.heatmap(var,vmin=0,vmax=10,annot=True,annot_kws=y,linewidth=10,linecolor="y")
```

```python
y={"fontsize":20,"color":"r"}
sns.heatmap(var,vmin=0,vmax=10,annot=True,annot_kws=y,linewidth=10,cbar=False)
```

```python
y={"fontsize":20,"color":"r"}
sns.heatmap(var,vmin=0,vmax=10,annot=True,annot_kws=y,linewidth=10,xticklabels=False)
plt.show()
```

```python
y={"fontsize":20,"color":"r"}
sns.heatmap(var,vmin=0,vmax=10,annot=True,annot_kws=y,linewidth=10,xticklabels=False,yticklabels=False)
plt.show()
```

```python
y={"fontsize":20,"color":"r"}
v=sns.heatmap(var,vmin=0,vmax=10,annot=True,annot_kws=y,linewidth=10,xticklabels=False,
            yticklabels=False)
v.set(xlabel="python",ylabel="pramod")
plt.show()
```

```python
y={"fontsize":20,"color":"r"}
v=sns.heatmap(var,vmin=0,vmax=10,annot=True,annot_kws=y,linewidth=10,xticklabels=False,
            yticklabels=False)
v.set(xlabel="python",ylabel="pramod")
sns.set(font_scale=10)
plt.show()
```
