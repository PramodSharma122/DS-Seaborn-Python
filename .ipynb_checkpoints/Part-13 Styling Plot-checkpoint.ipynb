# Styling Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=pd.read_csv("tips.csv")
var
```

```python
sns.boxplot(y=var['total_bill'])
plt.show()
```

```python
sns.barplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set_style("dark")
sns.barplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.barplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.barplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set_style("ticks")
sns.barplot(x="day",y="total_bill",data=var)
plt.show()
```

# Removing Axes Spines

```python

sns.set_style("white")
sns.barplot(x="day",y="total_bill",data=var)
sns.despine()
plt.show()
```

```python

sns.set_style("white")
plt.figure(figsize=(12,3))
sns.barplot(x="day",y="total_bill",data=var)
sns.despine()
plt.show()
```

```python

sns.set_style("white")
sns.set_context("poster",font_scale=2)
sns.barplot(x="day",y="total_bill",data=var)
sns.despine()
plt.show()
```

```python

sns.set_style("white")
sns.set_context("paper",font_scale=2)
sns.barplot(x="day",y="total_bill",data=var)

plt.show()
```

```python

sns.set_style("white")
sns.set_context("paper",font_scale=2)
sns.barplot(x="day",y="total_bill",data=var,palette="cool")

plt.show()
```
