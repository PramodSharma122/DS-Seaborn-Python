# Violin Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=pd.read_csv("tips.csv")
var
```

```python
sns.violinplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="time")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,linewidth=4)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,palette="Dark2")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=var,order=['Dinner','Lunch'])
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,palette="Dark2",saturation=0.1)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,color="y")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",hue="sex",data=var,split=True)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",hue="sex",data=var,split=True,scale="count")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=var,order=['Dinner','Lunch'],inner="quart")
plt.show()
```

```python
sns.violinplot(x=var["total_bill"])
plt.show()
```
