# Strip Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=pd.read_csv("tips.csv")
var
```

```python
sns.stripplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=2)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=2,edgecolor="r")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=2,
              edgecolor="r",jitter=5)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=2,
              edgecolor="r",size=10)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=2,
              edgecolor="r",size=10,marker="<")
plt.show()
```

```python

sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=2,
              edgecolor="r",alpha=0.2)
plt.show()
```

```python
sns.stripplot(x=var["total_bill"])
plt.show()
```
