# Box Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=pd.read_csv("tips.csv")
var
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="time")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",color="g")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",color="g",order=["Sun","Thur","Fri","Sat"])
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",color="g",order=["Sun","Thur","Fri","Sat"],
           showmeans=True)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,color="r",showmeans=True)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,color="w",showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"})
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,color="w",showmeans=True,
            meanprops={"marker":"+","markeredgecolor":"b"},linewidth=5)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,color="w",showmeans=True,
            meanprops={"marker":"+","markeredgecolor":"b"},palette="plasma")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(y="day",x="total_bill",data=var,color="w",showmeans=True,
            meanprops={"marker":"+","markeredgecolor":"b"},palette="plasma")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(data=var,color="w",showmeans=True,
            meanprops={"marker":"+","markeredgecolor":"b"},palette="plasma",orient="h")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x=var["total_bill"])
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(y=var["total_bill"])
plt.show()
```
