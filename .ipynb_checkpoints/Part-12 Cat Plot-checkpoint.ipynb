# Cat Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=pd.read_csv("tips.csv")
var
```

```python
sns.catplot(x="tip",y="size",data=var)
plt.show()
```

```python
sns.catplot(x="tip",y="size",data=var,hue="sex")
plt.show()
```

```python
sns.catplot(x="tip",y="size",data=var,hue="sex",palette="BuPu")
plt.show()
```

```python
sns.catplot(x="tip",y="size",data=var,hue="sex",palette="BuPu",height=3)
plt.show()
```

```python
sns.catplot(x="tip",y="size",data=var,hue="sex",palette="BuPu",kind="point")
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=var,hue="sex",palette="BuPu",kind="boxen")
plt.show()
```
