# Pair Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=pd.read_csv("tips.csv")
var
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var,hue="sex")
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex")
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",hue_order=["Male","Female"])
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",hue_order=["Male","Female"],palette="BuGn")
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Male","Female"],x_vars=["total_bill","tip"])
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Male","Female"],kind="reg")
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Male","Female"],kind="kde",diag_kind="hist")
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Male","Female"],markers=["*","<"])
plt.show()
```
