# Multiple Plots        (Facet Grid)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=pd.read_csv("tips.csv")
var
```

```python
fg=sns.FacetGrid(var)
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex")
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex",hue="day")
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex",hue="day")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="day",hue="sex")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex",hue="day")
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex",hue="day",palette="seismic")
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex",hue="day",palette="seismic")
fg.map(plt.bar,"total_bill","tip",edgecolor="g").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="time",hue="tip")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```
