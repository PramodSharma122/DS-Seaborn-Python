# Count Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=pd.read_csv("tips.csv")
var
```

```python
sns.countplot(x="sex",data=var)
plt.show()
```

```python
sns.countplot(x="sex",data=var,hue="smoker")
plt.show()
```

```python
sns.countplot(x="sex",data=var,hue="smoker",palette="bwr")
plt.show()
```

```python
sns.countplot(x="sex",data=var,hue="smoker",color="r")
plt.show()
```

```python
sns.countplot(x="sex",data=var,hue="smoker",color="r",saturation=0.1)
plt.show()
```
